```python
import tqdm
def calcular_velocidad_por_tramo(df, distancias_df):
    resultados = []
    for patente, grupo in tqdm.tqdm(df.groupby('patente')):
        grupo = grupo.sort_values(by='Fecha Hora')
        for i in range(len(grupo) - 1):
            origen = grupo.iloc[i]['Portico']
            destino = grupo.iloc[i + 1]['Portico']
            tiempo_origen = grupo.iloc[i]['Fecha Hora']
            tiempo_destino = grupo.iloc[i + 1]['Fecha Hora']
            tiempo_transcurrido = (tiempo_destino - tiempo_origen).total_seconds() / 3600  # convertir a horas
            
            distancia = distancias_df[(distancias_df['Origen'] == origen) & (distancias_df['Destino'] == destino)]['Distancia'].values
            if len(distancia) == 0:
                distancia = distancias_df[(distancias_df['Destino'] == origen) & (distancias_df['Origen'] == destino)]['Distancia'].values
            if len(distancia) > 0:
                distancia = distancia[0]
                velocidad = distancia / tiempo_transcurrido if tiempo_transcurrido > 0 else None
                resultados.append({
                    'patente': patente,
                    'origen': origen,
                    'destino': destino,
                    'distancia_km': distancia,
                    'tiempo_horas': tiempo_transcurrido,
                    'velocidad_kmh': velocidad
                })
    return pd.DataFrame(resultados)

# # Calcular velocidades por tramo
# velocidades_df = calcular_velocidad_por_tramo(df_new, distancias_df)

# # Mostrar los resultados
# print(velocidades_df)
```

```python
import pandas as pd

def calcular_velocidad_por_tramo_optimizado(df, distancias_df):
    # Hacer una copia del DataFrame original para no modificarlo
    df_copy = df.copy()

    # Convertir "Fecha Hora" a datetime si no lo está
    df_copy['Fecha Hora'] = pd.to_datetime(df_copy['Fecha Hora'])

    # Eliminar registros con patente vacía o con longitud menor a 5 caracteres
    df_copy = df_copy[df_copy['patente'].apply(lambda x: len(x) >= 5)]

    # Ordenar por patente y fecha/hora
    df_copy = df_copy.sort_values(by=['patente', 'Fecha Hora'])

    # Usar shift para obtener la fila anterior por grupo de patente
    df_copy['origen'] = df_copy['Portico']
    df_copy['destino'] = df_copy.groupby('patente')['Portico'].shift(-1)
    df_copy['tiempo_origen'] = df_copy['Fecha Hora']
    df_copy['tiempo_destino'] = df_copy.groupby('patente')['Fecha Hora'].shift(-1)

    # Calcular el tiempo transcurrido en horas
    df_copy['tiempo_transcurrido'] = (df_copy['tiempo_destino'] - df_copy['tiempo_origen']).dt.total_seconds() / 3600 

    # Merge con distancias_df para obtener la distancia entre origen y destino 
    merged_df = pd.merge(df_copy, distancias_df, how='left', left_on=['origen', 'destino'], right_on=['Origen', 'Destino'])

    # Filtrar filas donde la distancia es NaN
    merged_df = merged_df.dropna(subset=['Distancia'])

    # Calcular la velocidad solo si la distancia está presente
    merged_df['velocidad_kmh'] = merged_df['Distancia'] / merged_df['tiempo_transcurrido']

    # Filtrar filas donde destino es NaN (última fila por grupo después del shift)
    resultados_df = merged_df.dropna(subset=['destino'])

    # Seleccionar columnas relevantes para el resultado
    resultados_df = resultados_df[['patente', 'origen', 'destino', 'Distancia', 'tiempo_origen', 'tiempo_destino', 'tiempo_transcurrido', 'velocidad_kmh']]

    return resultados_df

# Datos de ejemplo de pórticos y distancias
distancias_data = {
    'Origen': ['Vespucio', 'Rinconada', 'Padre Hurtado', 'Malloco', 'Talagante', 'El Paico', 'Pomaire/Melipilla',
               'Rinconada', 'Padre Hurtado', 'Malloco', 'Talagante', 'El Paico', 'Pomaire/Melipilla', 'Puangue'],
    'Destino': ['Rinconada', 'Padre Hurtado', 'Malloco', 'Talagante', 'El Paico', 'Pomaire/Melipilla', 'Puangue',
                'Vespucio', 'Rinconada', 'Padre Hurtado', 'Malloco', 'Talagante', 'El Paico', 'Pomaire/Melipilla'],
    'Distancia': [3.8, 8.2, 6, 6.8, 7.9, 20.6, 20.7,
                  3.8, 8.2, 6, 6.8, 7.9, 20.6, 20.7]
}

distancias_df = pd.DataFrame(distancias_data)

# Suponemos que df_new es tu DataFrame existente con los datos de medición
# df_new = ...

# Calcular velocidades por tramo sin modificar df_new original
resultados_df_v2 = calcular_velocidad_por_tramo_optimizado(df_new, distancias_df)

# Mostrar los resultados
print(resultados_df_v2)
resultados_df_v2.to_excel
```
